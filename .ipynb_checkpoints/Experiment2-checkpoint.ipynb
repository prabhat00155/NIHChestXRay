```python
import pandas as pd
import numpy as np
import tensorflow as tf
import os
import matplotlib.pyplot as plt
import cv2
from sklearn.model_selection import train_test_split
```

```python
ROOT_FOLDER = '/datadisk/datasets/NIH_Chest_XRay'
IMAGE_FOLDER = '{}/images/'.format(ROOT_FOLDER)
DATA_FILE = '{}/Data_Entry_2017.csv'.format(ROOT_FOLDER)
TRAINING_FILE = '{}/train_val_list.txt'.format(ROOT_FOLDER)
TEST_FILE = '{}/test_list.txt'.format(ROOT_FOLDER)
```

```python
xray_df = pd.read_csv(DATA_FILE)
```

```python
xray_df.head()
```

```python
xray_df['Image Path'] = IMAGE_FOLDER + xray_df['Image Index']
```

```python
xray_df.head()
```

```python
label_list = ['Atelectasis',
 'Cardiomegaly',
 'Consolidation',
 'Edema',
 'Effusion',
 'Emphysema',
 'Fibrosis',
 'Hernia',
 'Infiltration',
 'Mass',
 'No Finding',
 'Nodule',
 'Pleural_Thickening',
 'Pneumonia',
 'Pneumothorax']
```

```python
for label in label_list:
    xray_df[label] = xray_df['Finding Labels'].str.split('|').apply(lambda x, y: int(y in x), args=(label,))
```

```python
xray_df.head()
```

```python
freq = list(map(lambda x: np.sum(xray_df[x]), label_list))
plt.bar(label_list, freq)
plt.xticks(rotation='vertical')
plt.xlabel('Diseases')
plt.ylabel('number of cases')
plt.show()
```

```python
plt.pie(freq, labels=label_list, radius=2, rotatelabels=True)
plt.show()
```

```python
xray_df['disease OHE'] = [np.array([xray_df.iloc[i][x] for x in label_list], dtype=np.uint8) for i in range(len(xray_df))]
```

```python
xray_df.head()
```

```python
xray_df['isTrain'] = [True for x in range(len(xray_df))]
```

```python
xray_df.loc[xray_df['Image Index'] == '00000001_000.png', 'isTrain'] = True
```

```python
with open(TEST_FILE) as f:
    for line in f:
        xray_df.loc[xray_df['Image Index'] == line.strip('\n'), 'isTrain'] = False
```

```python
xray_df.head()
```

```python
training_df = xray_df[xray_df['isTrain'] == True]
test_df = xray_df[xray_df['isTrain'] == False]
```

```python
def getNextBatch(starting_index, batch_size):
    data = [cv2.imread(xray_df.iloc[i]['Image Path']) 
            for i in range(starting_index, min(starting_index + batch_size, len(xray_df)))]
    label = xray_df.iloc[starting_index: starting_index + batch_size]['disease OHE']
    return data, label, starting_index + batch_size
```

```python
len(training_df)
```

```python
train_df, val_df = train_test_split(training_df, test_size=0.2)
```
